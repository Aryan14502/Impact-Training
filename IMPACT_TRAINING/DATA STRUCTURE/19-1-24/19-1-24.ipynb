```python
#Quick sort

def partition(arr,n,low,high):
    if(low<high):
        partition = quickSort(arr,n,low,high)
        quickSort(arr,n,low,partition-1)
        quickSort(arr,n,partition+1,high)
        
def quickSort(arr,n,low,high):
    pivot = arr[low]
    start=low
    end=high
    while(start<end):
        while(arr[start]<pivot):
            start+=1
            
        while(arr[end]>pivot):
            end-=1


        if low<high:
            arr[low],arr[high]=arr[high],arr[low]
    
    arr[low],arr[end]=arr[end],arr[low]
    return end
    
        

arr = list(map(int, input().split()))
n = len(arr)
low = 0
high = len(arr)-1
partition(arr,n,0,len(arr)-1)
```

```python
#Searching

#jump search
import math
arr = list(map(int, input().split()))
#jump will always be square root of len(arr)
ju = (math.sqrt(len(arr)))
jumpSize = int(ju)
key = int(input())

for i in range(0,len(arr),jumpSize):
    if arr[i]==key:
        print(f"found {arr[i]} at index : {i}")
    elif arr[i]>key:
        for j in range(i-jumpSize,key):
            if arr[j]==key:
                print(f"found {arr[j]} at index {j}")
    elif arr[i]<key:
        i += jumpSize    

```

```python
#Exponential search

def binarySearch(arr,low,high,key):
    while low<=high:
        mid=low+(high-low)//2
        if arr[mid]==key:
            return mid 
        elif arr[mid]>key:
            low=mid+1
        else:
            high = mid-1
    return -1

def exponentialSearch(arr,key,n):
    i = 1 
    if arr[0]==key:
        return 0
    while i<n:
        if arr[i]==key:
            return i    
        elif arr[i]<key:
            prev = i     
            i*=2
        else:
            binarySearch(arr,prev,i,key)
            
arr = list(map(int, input().split()))
n = len(arr)
key = eval(input())
exponentialSearch(arr,key,n)
```

```python
def print_custom_pattern(rows):
    current_number = 1

    for i in range(rows):
        for j in range(i + 1):
            if j == 0:
                print(current_number, end=" ")
            else:
                if i % 2 != 0:
                    print(current_number + i - 2 * j, end=" ")
                else:
                    print(current_number + i + 2 * j, end=" ")
            current_number += 1
        print()

# Specify the number of rows in the custom pattern
num_rows = 5

# Call the function to print the pattern
print_custom_pattern(num_rows)
```
